Now we iterate over the tokens and find the pair which is repeating.

```python
def get_stats(ids):   #tkns is the list of integers (the token ones probably!!)
    counts = {}        #this dic contains token pairs as keys , number of repeatation as value
    for pair in zip(ids , ids[1:]):
        counts[pair] = counts.get(pair,0) + 1
    return counts

stats = get_stats(tokens)
print(stats)
print(sorted(((v,k) for k,v in stats.items()) , reverse = True))
```

```python
top_pair = max(stats , key=stats.get)
top_pair
```

```python
def merge(ids,pair , idx):
    # in the list of ints (ids) , replace all consecutive occurences of pair with the new token idx
    newids = []
    i = 0
    while i < len(ids):
        # if we are not at the very last position AND the pair matches , replace it
        if i < len(ids) - 1 and ids[i] == pair[0] and ids[i+1] == pair[1]:
            newids.append(idx)
            i += 2
        else:
            newids.append(ids[i])
            i += 1
    return newids
```
